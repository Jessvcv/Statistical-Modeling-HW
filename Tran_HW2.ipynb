```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statistics import mode
from scipy.stats import norm as norm
```

```python
#Training Data
X1 = np.array([0,2,0,0,-1,1])
X2 = np.array([3,0,1,1,0,1])
X3 = np.array([0,0,3,2,1,1])
Y = np.array(["Red", "Red", "Red", "Green", "Green", "Red"])
```

```python
#the following test point is defined as test_point = [X1, X2, X3]
#we want to predict Y ("Green" or "Red")
test_point_X1 = 0
test_point_X2 = 0
test_point_X3 = 0
test_point = np.array([test_point_X1, test_point_X2, test_point_X3])
```

# Q1 : (1 point)
Complete the function to find the distance from the test_point coordinates to the data coordinates.

```python
def find_distance(test_point_X1, test_point_X2, test_point_X3, data_X1, data_X2, data_X3):
  distance = np.sqrt((test_point_X1 - data_X1)**2 + (test_point_X2 - data_X2)**2 + (test_point_X3 - data_X3)**2)
  return distance
```

# Q3. (1 point)
Take the code from Q2. and package it into a function called `find_all_distances` which takes in the test point and the data and returns an array of distances.


```python
def find_all_distances(test_point_X1, test_point_X2, test_point_X3, X1, X2, X3):
  # initialize a numpy array of zeros with the same length as the training data
    distances_array = np.zeros(len(X1))

    # loop over all data points
    for i in range(len(X1)):
        # calculate distance from test point to ith training point
        dist = find_distance(test_point_X1, test_point_X2, test_point_X3,
                             X1[i], X2[i], X3[i])
        # store in array
        distances_array[i] = dist

    return distances_array
```

# Q4. (1 point)
Write detailed comments for each line in the code below. What is it doing?

Rename the function and variable names so that it is more descriptive.

```python
#test_point: array representing the test point
#X1, X2, X3: array representing feature values of the training data
#Y: array of class labels that correspond to training data
#k: an odd integer specifying the number of nearest neighbors to consider
def knn_classifier(test_point, X1, X2, X3, Y, k):

  #computes distances between test point and all training points
  distances = find_all_distances(test_point[0], test_point[1], test_point[2], X1, X2, X3)

  #get the indices that sorts the distances array from smallest to largest
  sorted_indices = distances.argsort()

  #select the indices of the k closest neighbor
  k_nearest_indices = sorted_indices[:k]

  #returns most frequent labels
  predicted_label = mode(Y[k_nearest_indices])
  return predicted_label
```

# Q8. (2 points)
We want to create a new array which has labels `"hot"`, `"cold"`, `"okay" `which correspond to the temperatures in `temp_array` . `label_list` should be the same length as `temp_array` and values should correspond as follows:

*   `< 50 `should be labeled `"cold"`
*   `> 85` should be labeled `"hot" `
*   and values between will be `"okay"`


```python
temp_array = np.array([12, 35, 69, 90, 35, 77, 34, 90, 40, 67, 76, 80, 88])
label_list = []
```

```python
for temp in temp_array:
    if temp < 50:
        label_list.append("cold")
    elif temp > 85:
        label_list.append("hot")
    else:
        label_list.append("okay")

print(label_list)
```
